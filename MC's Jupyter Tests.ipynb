```python
import matplotlib
import pandas as pd
```

```python
colleges = ["Christ's", "Hughes Hall", "Sidney", "Fitzwilliam", "Trinity", "Peterhouse", "Girton", "Selwyn", "Magdalene", "Pembroke"]
location = ["Central", "Outside", "Very far"]
size = ["Small", "Medium", "Large"]
```

```python
data = {'College': colleges,
        'Location': [location[i] for i in [0,1,0,1,0,0,2,1,1,0]],
       'Size': [size[i] for i in [1,0,1,2,2,0,1,1,1,0]]}
```

```python
c_df = pd.DataFrame(data)
```

```python
c_df
```

```python
c_df = pd.DataFrame(data, columns = ['College', 'Location', 'Size', 'Age'])
```

```python
c_df
```

```python
age = ['Old', 'Modern']
c_df['Age'] = pd.Series(age[0], index=[0,2,4,5,6,7,8,9])
```
